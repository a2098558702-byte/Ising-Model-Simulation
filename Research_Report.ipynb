```python
#  Ising  

from ising_core import IsingSimulation
import matplotlib.pyplot as plt

#  50x50 
sim = IsingSimulation(N=50, J=1.0)

history = sim.run_simulation(T=2.27, steps=2000)

plt.plot(history)
plt.title("Magnetization vs Time")
plt.xlabel("Steps")
plt.ylabel("Magnetization")
plt.show()
```
